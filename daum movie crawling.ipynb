```python
import re
import requests
import pickle
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome('chromedriver')
```

```python
def download_image(url,name):
    img_data = requests.get(url).content
    with open(f'poster/daum/{name}.jpg', 'wb') as handler:
        handler.write(img_data)
```

```python
i=0
patience = 0
result = {}
while True:
    if patience==100:break
    driver.get(f'https://movie.daum.net/moviedb/main?movieId={i}')
    try:
        genres = driver.find_element_by_xpath('//*[@id="mArticle"]/div[2]/div[2]/div[1]/div[1]/div[2]/div/dl[1]/dd[1]').text.split('/')
        img_url = driver.find_element_by_tag_name('img').get_attribute('src')
        print(genres)
        print(img_url)
        i+=1
    except:
        patience+=1
        i+=1
        continue
    if len(genres)!=0 and len(img_url)!=0:
        result[len(result)]=genres
        download_image(img_url,str(len(result)-1))
        patience=0
```
