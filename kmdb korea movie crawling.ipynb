```python
import re
import requests
import pickle
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome('chromedriver')
driver.get('https://www.kmdb.or.kr/login')
driver.implicitly_wait(3)
```

```python
def get_genre(driver):
    page_source = driver.page_source
    if '장르' not in driver.page_source:
        return None
    i=3
    while True:
        if i==7:return None
        try:
            genres = driver.find_element_by_xpath(f'//*[@id="fieldset"]/div[{i}]/div[4]/dl/dd[1]').text.split(' ')
        except:
            i+=1
            continue
        break
    return genres
def get_image_url(driver):
    page_source = driver.page_source
    temp1 = page_source.find('.jpg')
    if temp1==-1:return None
    temp2 = page_source[temp1-100:temp1].find('http')+temp1-100
    poster_url = page_source[temp2:temp1+4]
    if 'http' not in poster_url:return None
    if 'still' in poster_url:return None
    return poster_url

def download_image(url,name):
    img_data = requests.get(url).content
    with open(f'poster/kmdb_korea/{name}.jpg', 'wb') as handler:
        handler.write(img_data)
```

```python
patience = 0
k_result = {}
for i in range(10000,99999):
    if patience==100:break
    country = 'K'
    index = '0'*(5-len(str(i)))+str(i)
    driver.get(f'https://www.kmdb.or.kr/db/kor/detail/movie/{country}/{index}')
    driver.implicitly_wait(3)
    
    genres = get_genre(driver)
    image_url = get_image_url(driver)
    print(genres)
    print(image_url)
    if (genres==None) and (image_url==None):
        patience+=1
        continue
    else:
        if (genres!=None) and (image_url!=None):
            k_result[len(k_result)] = genres
            download_image(image_url,str(len(k_result)-1))
        patience=0
```

```python
with open('poster/kmdb_korea.txt', 'wb') as f:
    pickle.dump(k_result, f)
```
